```python
import psycopg2
from psycopg2 import sql
import csv
import os

# Database connection parameters
DB_CONFIG = {
    'dbname': 'final_project',  # Replace with your database name
    'user': 'postgres',         # Replace with your database username
    'password': 'Vijay@_2002',   # Replace with your database password
    'host': 'localhost',        # Replace with your host
    'port': 5432                # Replace with your PostgreSQL port (default is 5432)
}

# Directory to save the CSV files
RAW_FOLDER = 'raw'

# Ensure the directory exists
os.makedirs(RAW_FOLDER, exist_ok=True)

def save_data_to_csv(data, columns, filename):
    """Save data to a CSV file."""
    file_path = os.path.join(RAW_FOLDER, filename)
    
    # Write data to the CSV file
    with open(file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(columns)  # Write the header
        writer.writerows(data)    # Write the data

    print(f"Data saved to {file_path}")

def fetch_and_save_data(query, filename, columns):
    """Fetch data from the database and save it to a CSV file."""
    try:
        conn = psycopg2.connect(**DB_CONFIG)
        cursor = conn.cursor()

        # Execute the query
        cursor.execute(query)

        # Fetch all results
        data = cursor.fetchall()

        # Save the data to a CSV file
        save_data_to_csv(data, columns, filename)

    except psycopg2.Error as e:
        print(f"Error connecting to the database: {e}")
    finally:
        # Close the cursor and connection
        if cursor:
            cursor.close()
        if conn:
            conn.close()

# Fetch and save data from the "User" table
user_query = sql.SQL("SELECT id, name, designation, role FROM public.\"User\"")
user_columns = ['id', 'name', 'designation', 'role']
fetch_and_save_data(user_query, 'users.csv', user_columns)

# Fetch and save data from the "Course" table
course_query = sql.SQL("SELECT id, title, duration, difficulty_level FROM public.\"Course\"")
course_columns = ['id', 'title', 'duration', 'difficulty Level']
fetch_and_save_data(course_query, 'courses.csv', course_columns)

# Fetch and save data from the "Certificates" table
certificates_query = sql.SQL("SELECT id, is_certified, user_id,assignment_id  FROM public.\"Certificates\"")
certificates_columns = ['id', 'is_certified', 'user_id','assignment_id']
fetch_and_save_data(certificates_query, 'certificates.csv', certificates_columns)

# Fetch and save data from the "LearningPath" table
learning_path_query = sql.SQL("SELECT id, title, description FROM public.\"LearningPath\"")
learning_path_columns = ['id', 'title', 'description']
fetch_and_save_data(learning_path_query, 'learning_paths.csv', learning_path_columns)

# Fetch and save data from the "CourseAssignment" table
course_assignment_query = sql.SQL("SELECT id, user_id, course_id, progress, assignment_date, quiz_score,engagement_score,assignment_grade FROM public.\"CourseAssignment\"")
course_assignment_columns = ['id', 'user_id', 'course_id', 'progress', 'assignment_date', 'quiz_score','engagement_score', 'assignment_grade']
fetch_and_save_data(course_assignment_query, 'course_assignments.csv', course_assignment_columns)

# Fetch and save data from the "PerformanceRating" table
performance_rating_query = sql.SQL("SELECT id, rating, user_id, assignment_id FROM public.\"PerformanceRating\"")
performance_rating_columns = ['id', 'rating', 'user_id', 'assignment_id']
fetch_and_save_data(performance_rating_query, 'performance_ratings.csv', performance_rating_columns)

# Fetch and save data from the "PerformanceSummary" table
performance_summary_query = sql.SQL("SELECT id, user_id, average_rating, learning_path_id FROM public.\"PerformanceSummary\"")
performance_summary_columns = ['id', 'user_id', 'average_rating', 'learning_path_id']
fetch_and_save_data(performance_summary_query, 'performance_summaries.csv', performance_summary_columns)

# Fetch and save data from the "CourseLearningPath" table
course_learning_path_query = sql.SQL("SELECT id, course_id, learning_path_id FROM public.\"CourseLearningPath\"")
course_learning_path_columns = ['id', 'course_id', 'learning_path_id']
fetch_and_save_data(course_learning_path_query, 'course_learning_paths.csv', course_learning_path_columns)

# You can continue to add more tables as needed.
```
